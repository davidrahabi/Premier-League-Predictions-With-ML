```python
#clean data for machine learning
matches.dtypes

#dtypes are the type of data in each column
#ML can only work with numbers, not objects
#we need to convert the object types into a numerical type of value
```

```python
#turn date object into a date-time number
matches["date"] = pd.to_datetime(matches["date"])
#converts existing date column to a pandas datetime object and overwrites it
matches.dtypes #now we can easily make predictors using the date column
```

```python
#create basic predictors to build off of and add complexity to
#create value for Home/Away advantage

matches["venue_code"] = matches["venue"].astype("category").cat.codes
#convert the "home" and "away" strings into categories, then convert the 
#categories into numerical codes
# 0 = away  1 = home
```

```python
#create value for the opponent squad
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
#every opponent now has its own unique code
```

```python
#hour column based on time of day -- certain teams may play better at certain times of the day
matches["hour"] = matches["time"].str.replace(":.+", "", regex = True).astype("int")
#replace 16:30 with 16 -- remove colon and minutes
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
#set up target that we can try to predict
matches["target"] = (matches["result"] == "W").astype("int")
        #if its win it returns true. astype will convert it to a 1
#our target is whether the team won or not
#loss or draw = 0
#win = 1
```

```python
#initial ML model
import sklearn
from sklearn.ensemble import RandomForestClassifier

# this model will pick up nonlinearities in the data
#understands that certain values like opp_code are arbitrary values non-indicative of difficulty
#linear models can't do this

rf = RandomForestClassifier(min_samples_split = 10,n_estimators=50, random_state = 1)
#a random forest is s series of decision trees, but each decision tree has slightly different parameters
#n_estimators is the amt of trees
#min_sample_split is the num of samples we want in a leaf of a tree before we split to the next node
#random forest has many random parameters in it, so random_state=1 means results will be the same as long as data is the same

```

```python
train = matches[matches["date"] < '2024-01-01']
 #splitting up training and test data
 #this is time series data, we must make sure that all data in the test set comes after the data in the training set
#have to make sure you are using data from the past to predict the future

test = matches[matches["date"] > '2024-01-01']
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]

rf.fit(train[predictors],train["target"])
#trains random forest model use list of predictors to predict the target column
```

```python
preds = rf.predict(test[predictors])
#generate predictions, pass in test data
preds
```

```python
#determine accuracy of model
from sklearn.metrics import accuracy_score

acc = accuracy_score(test["target"], preds)
acc
```

```python
#check which situations accuracy was higher vs lower
#create dataframe

combined = pd.DataFrame(dict(actual = test["target"], prediction = preds))
#combines actual valuse and predicted values
#first arg is actual, second arg is prediction vals

pd.crosstab(index = combined["actual"], columns = combined["prediction"])
#two way tablet that will show us what happened when we preicted a 0 or 1
#shows the amount of times we were correct -- (0,0) is amt of times L prediciton was correct and (1,1) v.v

#issue: we want to predict wins correctly, but it seems that we are much better at predicting losses compared to the wins
```

```python
#we will use precision score -- when we predicted a win, what percentage of the time did the team actually win
from sklearn.metrics import precision_score
precision_score(test["target"], preds)
#we only had a 56% precision when predicting a win
```

```python
#improving precision with rolling averages -- creating more predictions
#compute rolling averages for each team based upon most recent games
#create one dataframe for every squad

grouped_matches = matches.groupby("team")
group = grouped_matches.get_group("Manchester City")

#compute rolling averages from previous 5 match weeks
#takes in team, a set of cols that we want to compute rolling averages for, and a group of new cols to assign rolling averages to
def rolling_averages(group, cols, new_cols):
    #sort group by date
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(5,closed = 'left').mean()
     #closed will stop the function from including current week (the one we are trying to predict/assign the average to)
    group[new_cols] = rolling_stats #assign rolling stats to original dataframe, but with new names
    #drop any missing values/matchweeks
    group = group.dropna(subset = new_cols)
    return group

    

```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
#creating rolling columsn for each predictor we are using

```

```python
rolling_averages(group,cols,new_cols)
#now we can pass these rolling values to improve our predictions

```

```python
matches_rolling = matches.groupby("team").apply(lambda x :rolling_averages(x,cols,new_cols))
#apply the function to all matches -- ^^ applies rolling_averages function to each team
#creates different dataframes for each team then applies rolling averages

matches_rolling = matches_rolling.droplevel('team') #just removing the team names that were extra row index levels

matches_rolling.index = range(matches_rolling.shape[0]) #there are duplicate indices between the dataframes, this makes all of them unique
matches_rolling
```

```python
#retrain model based on new predictions

def make_predictions(data, predictors):
    train = data[data["date"] < '2024-01-01']
    test = data[data["date"] > '2024-01-01']
    rf.fit(train[predictors],train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual = test["target"], predicted = preds), index = test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision #all of this was done previously, we are now just putting in into a function

    
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
precision
#precision went up by about 1%, probably makes sense considering it was much higher originally than in the video
```

```python
#combine home and away predictions
#ex: we can accidentally predict that both teams will win because we are making two predictions for both teams for each game

#team name is different than opponent name -- need to make them the same

class MissingDict(dict):
    __missing__ = lambda self, key: key
map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", 
"Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves", 
"Sheffield United": "Sheffield Utd",  "Nottingham Forest":"Nott'ham Forest", "West Bromwich Albion": "West Brom"} 

mapping = MissingDict(**map_values)
   
```

```python
combined["new_team"] = combined["team"].map(mapping) #creates new team field in combined
#merge dataframe with itself
merged = combined.merge(combined, left_on = ["date", "new_team"], right_on=["date", "opponent"])
merged #now we can see both sides of the prediction, so we can see where the predictions actually lined up
```
